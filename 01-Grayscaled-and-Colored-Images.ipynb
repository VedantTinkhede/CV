```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image in BGR format and convert it to RGB for correct Matplotlib display
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Load the image in grayscale mode
img_grayscale = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png", 0)
print(img.shape)  # Print the shape of the image (height, width, channels)
plt.imshow(img_grayscale, cmap="gray")  # Display the grayscale image
plt.show()  

# Load the image again and display only the green channel
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
plt.imshow(img[:, :, 1], cmap="gray")  # Show the green channel as a grayscale image
plt.show()  

# Remove green and red channels, keeping only the blue channel
img[:, :, 1] = 0  # Set green channel to zero
img[:, :, 2] = 0  # Set red channel to zero
plt.imshow(img)  # Display the modified image
plt.show()  

# Remove blue and red channels, keeping only the green channel
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
img[:, :, 2] = 0  # Set red channel to zero
img[:, :, 0] = 0  # Set blue channel to zero
plt.imshow(img)  # Display the modified image
plt.show()  

# Resize the image to 200x200 pixels and display it
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
img[:, :, 0] = 0  # Remove blue channel
img[:, :, 1] = 0  # Remove green channel
resized_img = cv2.resize(img, (200, 200))  # Resize image to 200x200
plt.imshow(resized_img)  # Display the resized image
plt.show()  

# Mix red and green channels and store the result in the red channel
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
mixed_channel = cv2.add(img[:, :, 2], img[:, :, 1])  # Add red and green channels
img[:, :, 2] = mixed_channel  # Store the sum in the red channel
plt.imshow(img)  # Display the modified image
plt.show()  

# Mix blue and green channels and store the result in the green channel
img = cv2.imread(r"C:\Users\ACER\Desktop\images\rcoem.png")
mixed_channel = cv2.add(img[:, :, 0], img[:, :, 1])  # Add blue and green channels
img[:, :, 1] = mixed_channel  # Store the sum in the green channel
plt.imshow(img)  # Display the modified image
plt.show()  
```
