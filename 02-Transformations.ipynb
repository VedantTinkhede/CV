```python
# -*- coding: utf-8 -*-
"""
CV_30/1/25.ipynb - Modified for Jupyter Notebook

This script performs various image transformations including translation, rotation, 
scaling, shearing, and reflection using OpenCV.
"""

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images in Jupyter Notebook
def show(img, title="Image"):
    """Displays an image using Matplotlib."""
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
    plt.title(title)
    plt.axis("off")  # Hide axis
    plt.show()

# Load the image (Ensure the image exists in the correct path)
image_path = (r"C:\Users\ACER\Downloads\Samsugn-QN90B2.jpg" ) # Update the path if needed
img = cv2.imread(image_path)

if img is None:
    print("Error: Image not found. Check the file path.")
else:
    show(img, "Original Image")

    # ----------------- Translation -----------------
    tx, ty = 100, 50  # Translate 100 pixels right and 50 pixels down
    translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
    translated_img = cv2.warpAffine(img, translation_matrix, (img.shape[1], img.shape[0]))
    show(translated_img, "Translated Image")

    # ----------------- Rotation -----------------
    angle = 45  # Rotate by 45 degrees
    centre = (img.shape[1] // 2, img.shape[0] // 2)  # Center of the image
    rotation_matrix = cv2.getRotationMatrix2D(centre, angle, 1)  # Scale factor = 1
    rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
    show(rotated_img, "Rotated Image")

    print("Rotation Matrix:\n", rotation_matrix)

    # ----------------- Scaling -----------------
    scaled_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR)  # Scale down by 50%
    show(scaled_img, "Scaled Image (50%)")

    # ----------------- Shearing -----------------
    shear_matrix = np.float32([[1, 0.2, 0], [0.2, 1, 0]])  # Shearing transformation matrix
    sheared_img = cv2.warpAffine(img, shear_matrix, (img.shape[1], img.shape[0]))
    show(sheared_img, "Sheared Image")

    # ----------------- Reflection (Flipping Horizontally) -----------------
    reflection_matrix = np.float32([[-1, 0, img.shape[1]], [0, 1, 0]])  # Horizontal reflection
    reflected_img = cv2.warpAffine(img, reflection_matrix, (img.shape[1], img.shape[0]))
    show(reflected_img, "Reflected Image (Horizontal Flip)")
```
